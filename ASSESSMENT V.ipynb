```python
#Importing Relevant Python  Libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Loading the csv file

CarSales=pd.read_csv('Car-sales.csv')
```

```python
#Showing the loaded csv file

CarSales
```

# Understanding the Dataset/Data Cleaning

```python
#Question 1 
#Identifying the Number of Rows and Columns in the dataset

Number_of_Rows= CarSales.shape[0]
Number_of_Columns= CarSales.shape[1]
```

```python
#Printing the Number of Rows and Columns in the dataset

print(f"Number of Rows: {Number_of_Rows}")
print(f"Number of Columns: {Number_of_Columns}")
```

```python
#Question 2 
#Identifying the Data Type of each column

CarSales.dtypes
```

```python
#Question 3
#Counting any missing values in the dataset

CarSales.isna().sum()
```

```python
# Filling the missing values in the 'year_resale_value' column with the it's average value (Hint: Find the Average of the column first)

Avg=CarSales['__year_resale_value'].mean()

Avg
```

```python
#Replacing the Missing Values with the average calculated (18.072975206611574)

CarSales['__year_resale_value'].replace(np.nan, Avg, inplace=True )
```

```python
#Printing the dataset to see the changes

CarSales
```

```python
#Dropping other rows with missing values

CarSales.dropna(inplace=True)
```

```python
#Displaying the dataset to see the changes

CarSales
```

```python
#Question 4 
#Cleaning Column four (4) to remove the underscores (Hint:Rename the Column)

CarSales.rename(columns={'__year_resale_value':'Year Resale Value'}, inplace=True)
```

```python
#Display the dataset to see the effect

CarSales
```

```python
#Describing the dataset
#Carrying out some measures of central tendency on the numerical columns(Hint: Find the Mean, Median and Standard Deviation)

CarSales.describe()
```

```python
#Question 5
#Checking for duplicates
CarSales.duplicated()
```

# SPECIFIC ANALYSIS

```python
#6.Checking for the manufacturer with the highest and lowest average sales volume

Average_sales_volume=CarSales.groupby('Manufacturer')['Sales_in_thousands'].mean()
Average_sales_volume
```

```python
#Highest Sales Volume
CarSales.groupby('Manufacturer')['Sales_in_thousands'].mean().idxmax()
```

```python
#Checking for the manufacturer with the lowest average sales volume

CarSales.groupby('Manufacturer')['Sales_in_thousands'].mean().idxmin()
```

```python
#7.Displaying the distribution of car prices in the dataset

pv3 = CarSales.pivot_table(index='Manufacturer', values='Price_in_thousands',aggfunc='sum')
pv3
pv3.plot(kind='hist', figsize=(10,8))
plt.xlabel('Manufacturer')
plt.ylabel('Sum of Prices')
plt.title('Distribution of Car Prices')
plt.show()
```

```python
#8. Correlation of Numerical Values in Dataset. 

CarSales_correlation=CarSales.corr()
CarSales_correlation
```

```python
#Plotting the Correlation matrix

plt.figure(figsize=(10, 8))
sns.heatmap(CarSales_correlation, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

```python
#9. Plotting the relationship between price and latest launch year.(Hint: Change Latest_Launch Data Type for linearity of data type)

CarSales.Latest_Launch= CarSales.Latest_Launch.astype('datetime64[ns]')
```

```python
#Relationship Between Price and Latest Launch

pv1 = CarSales.pivot_table(index='Latest_Launch', values='Price_in_thousands',aggfunc='sum')
pv1
pv1.plot(kind='line', figsize=(8,4))
plt.xlabel('Latest_Launch')
plt.ylabel('Price_in_thousands')
plt.title('Relationship of Car price and latest launch year')
plt.show()
```

```python
#Slicing from Year 2011 to 2013 for more insight into the data

after_2010=CarSales[CarSales['Latest_Launch'] > '01-01-2011']
after_2010
```

```python
#Question 10
#Grouping Cars By Manufacturers and their Prices

Average_Price=CarSales.groupby('Manufacturer')['Price_in_thousands'].mean()
Average_Price
```

```python
#Grouping Cars by Manufactures and their fuel efficiency

Average_Fuel_Efficiency=CarSales.groupby('Manufacturer')['Fuel_efficiency'].mean()
Average_Fuel_Efficiency
```

```python
#Question 11
#Visualizing the Most Popular car Brands

pv2 = CarSales.pivot_table(index='Manufacturer', values='Sales_in_thousands',aggfunc='sum')
pv2
pv2.plot(kind='bar', figsize=(8,4))
plt.xlabel('Manufacturer')
plt.ylabel('Sales_in_thousands')
plt.title('Most Popular Car Brand')
plt.show()

```

```python
#Question 12
#Comparing the Highest resale value with Car Model

Resale_Value=CarSales.groupby('Model')['Year Resale Value'].mean()
Resale_Value
```

```python
#Extracting the Model with the Highest Year Resale value

CarSales.groupby('Model')['Year Resale Value'].mean().idxmax()
```

```python
#Comparing the Model with the Highest Resale value with it's initial price

pv3 = CarSales.pivot_table(index='Model', values='Price_in_thousands',aggfunc='sum')
pv3
pv3.plot(kind='hist', figsize=(8,4))
plt.xlabel('Model')
plt.ylabel('Price_in_thousands')
plt.title('Resale value VS Initial Price')
plt.show()
```

```python
#Question 13
#Finding the top 3 fuel-efficient cars with an engine size greater than 2.5litres (Hint: Filter the variables concerned)

filtering= CarSales[CarSales['Engine_size']>2.5]
filtering
```

```python
#Using fuel efficiency, sort the top three
top_three= filtering.sort_values(by='Fuel_efficiency',ascending=False).head(3)
top_three
```

```python
#Showing the 'top three' model of cars with fuel efficiency

(top_three['Model'])
```

```python
#Question 14
#Identifying which Lexus Model has the highest horsepower(Hint:Filter the model in the dataset first)
#Filtering the dataset

lexus_model= CarSales[CarSales['Manufacturer']== 'Lexus']
lexus_model
```

```python
#finding the Lexus Model with the highest horsepower
#Showing the properties of the Lexus Model with the highest horsepower

highest_horsepower= lexus_model.loc[lexus_model['Horsepower'].idxmax()]
highest_horsepower

```

```python
#Extracting the name of the model with the highest horsepower
highest_horsepower['Model']
```
